```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
```

```python
import matplotlib.pyplot as plt
import random
```

```python
%matplotlib inline
```

```python
def gen_image_loss(epochs=32):
    image_number = 0
    while True:
        loss = []
        for x in range(epochs):
            loss.append((image_number * epochs + x, random.randint(0, 10)))
            
        image_number += 1
        yield loss
```

```python
colors = ['r', 'b', 'g']
```

```python
images = 5
loss_generator = gen_image_loss()
fig = plt.figure()
ax1 = fig.add_subplot(111)


for image in range(images):
    loss = next(loss_generator)
    x, y = zip(*loss)
    label = "Image {}".format(image)
    ax1.plot(x, y, c=colors[image % len(colors)], label=label)
    
plt.legend(loc='upper left');
plt.show()
```
